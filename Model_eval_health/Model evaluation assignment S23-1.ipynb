```python
"""
Read the data stored in your local machine https://www.kaggle.com/datasets/andrewmvd/fetal-health-classification
"""
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
df = pd.read_csv('fetal_health.csv')
df.head(10)
```

```python
df.info()
```

```python
X = df.drop(columns='fetal_health')
y = df['fetal_health']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=1)
```

### Perform classification task using 5 different models, and compare their accuracies (20 pts for each model)

#### Use the best hyperparameters returned from GridSearchCV to re-train the models, and compare the accuracies of all 5 models.
The followings include hyperparameters of mentioned models:
* Logistic Regression:
    * C: from 0.1 to 1, step = 0.3
    * multi_class: auto, ovr, multinomial
    * solver: newton-cg
* Decision Tree:
    * criterion: gini, entropy, log_loss
    * max_features: sqrt, log2
    * max_depth: 2 to 5
* KNN:
    * n_neighbors: 3 to 7
    * weights: uniform, distance
* SVC:
    * degree: 2 to 5
    * C: 0.1 to 1, step = 0.3
    * kernel: poly, rbf
* SGD:
    * loss: hinge, log_loss, modified_huber

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
# Logistic Regression
log_params = {
    'C': np.arange(0.1, 1.1, 0.3),
    'multi_class': ['auto', 'ovr', 'multinomial'],
    'solver': ['newton-cg']
}
log_grid = GridSearchCV(LogisticRegression(), log_params, cv=5)
log_grid.fit(X_train, y_train)
log_est = log_grid.best_estimator_
log_ypred = log_est.predict(X_test)
log_acc = accuracy_score(y_test, log_ypred)
print(f"Logistic Regression accuracy: {log_acc}")

```

```python
from sklearn.tree import DecisionTreeClassifier
# Decision Tree
dectree_params = {
    'criterion': ['gini', 'entropy'],
    'max_features': ['sqrt', 'log2'],
    'max_depth': range(2, 5)
}
dec_grid = GridSearchCV(DecisionTreeClassifier(), dectree_params, cv=4)
dec_grid.fit(X_train, y_train)
dec_best = dec_grid.best_estimator_
dec_ypred = dec_best.predict(X_test)
dec_acc = accuracy_score(y_test, dec_ypred)
print(f"Decision Tree accuracy: {dec_acc}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# KNN
knn_params = {
    'n_neighbors': range(3, 7),
    'weights': ['uniform', 'distance']
}
knn_grid = GridSearchCV(KNeighborsClassifier(), knn_params, cv=5)
knn_grid.fit(X_train, y_train)
knn_best = knn_grid.best_estimator_
knn_pred = knn_best.predict(X_test)
knn_acc = accuracy_score(y_test, knn_pred)
print(f"KNN accuracy: {knn_acc}")
```

```python
from sklearn.svm import SVC
# SVC
svc_params = {
    'degree': range(2, 5),
    'C': np.arange(0.1, 1.1, 0.3),
    'kernel': ['poly', 'rbf']
}
svc_grid = GridSearchCV(SVC(), svc_params, cv=6)
svc_grid.fit(X_train, y_train)
svc_best = svc_grid.best_estimator_
svc_pred = svc_best.predict(X_test)
svc_acc = accuracy_score(y_test, svc_pred)
print(f"SVC accuracy: {svc_acc}")
```

```python
from sklearn.linear_model import SGDClassifier
# SGD
sgd_params = {
    'loss': ['hinge', 'log', 'modified_huber']
}
sgd_grid = GridSearchCV(SGDClassifier(), sgd_params, cv=6)
sgd_grid.fit(X_train, y_train)
sgd_best = sgd_grid.best_estimator_
sgd_pred = sgd_best.predict(X_test)
sgd_acc = accuracy_score(y_test, sgd_pred)
print(f"SGD accuracy: {sgd_acc}")
```

```python
print(f"Logistic Regression accuracy: {log_acc}")
print(f"Decision Tree accuracy: {dec_acc}")
print(f"KNN accuracy: {knn_acc}")
print(f"SVC accuracy: {svc_acc}")
print(f"SGD accuracy: {sgd_acc}")
```
